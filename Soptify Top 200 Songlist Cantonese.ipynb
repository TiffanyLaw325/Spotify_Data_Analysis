```python
from gettext import npgettext
from importlib.metadata import requires
import requests 
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#import url
url = 'https://kworb.net/spotify/country/hk_daily.html'
#url = 'https://kworb.net/ww/index_full.html'
req = requests.get(url)
req.encoding = 'utf-8' # make traditional Chinese visible
req
# <Response [200]> meaning it is successful
print(req.text)
```

```python
soup = BeautifulSoup (req.text, features ='html.parser')
print(soup.prettify())   # places in a tidier style
print(soup.title.text)  #Spotify Daily Chart - Hong Kong
table1 = soup.find('table',id='spotifydaily') 
# Find table called spotifydaily
assert table1
```

```python
headers=[]
for th_cell in table1.find_all('td'):
  title = th_cell.text
  headers.append(title)
assert headers
```

```python
#headers = list(map(lambda x: x.text,table1.find_all('th')))
```

```python
spotify_data = pd.DataFrame (columns = headers)
```

```python
for row in table1.find_all('tr')[1:]:      #1
  cells = row.find_all('td')               #2
  cell_text = [ i.text for i in cells ]    #3
  pos = len(spotify_data)                  #4
  spotify_data.loc[pos] = cell_text        #5
```

```python
print(spotify_data)
```

```python
import datetime
from datetime import timedelta
yyyymmdd = datetime.datetime.today()
yyyymmdd = yyyymmdd-timedelta(days=2)
yyyymmdd = yyyymmdd.strftime('%Y%m%d')
spotify_data.to_csv(f"spotify_data_{yyyymmdd}.csv",index=False)
```
